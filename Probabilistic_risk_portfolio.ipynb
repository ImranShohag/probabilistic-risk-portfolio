import numpy as np

# Revenue Simulation

avg_fee = np.random.normal(200, 50, n)
revenue = clients * avg_fee

mean_rev = np.mean(revenue)

p5_rev = np.percentile(revenue, 5)
p95_rev = np.percentile(revenue, 95)

print("Mean Revenue:", mean_rev)

print("95th percentile Revenue:", p95_rev)

plt.hist(revenue, bins=50)
plt.axvline(mean_rev, color='red', label='Mean')

plt.legend()

# Operational Loss Simulation

```python
# Strategy Comparison
strategy_A = np.random.normal(10, 5, n)
strategy_B = np.random.normal(15, 15, n)

mean_A = np.mean(strategy_A)
mean_B = np.mean(strategy_B)

prob_neg_A = np.mean(strategy_A < 0)
prob_neg_B = np.mean(strategy_B < 0)

print("Strategy A Mean:", mean_A, "Probability Negative:", prob_neg_A)
print("Strategy B Mean:", mean_B, "Probability Negative:", prob_neg_B)
```

```python
# Scenario Stress Test (20% demand drop)
clients_stress = clients * 0.8
revenue_stress = clients_stress * avg_fee

mean_stress = np.mean(revenue_stress)
p5_stress = np.percentile(revenue_stress, 5)

print("Mean Stress Revenue:", mean_stress)
print("5th percentile Stress Revenue:", p5_stress)
```

# Executive Summary

Monte Carlo analysis shows that revenue has uncertainty due to demand and pricing.
The 5th percentile revenue is much lower than expected revenue, showing downside risk.

Operational losses have tail risk. 95th percentile loss is much higher than average. Liquidity buffer should cover this.

High-return strategies (Strategy B) have high risk. Low-risk strategy (A) is safer if survival is important.

Stress test (20% demand drop) shows revenue falls significantly, demand is main risk driver.

# Actionable Recommendations

- Keep liquidity buffer ≥ 95th percentile loss
- Monitor demand regularly
- Avoid high-risk strategies if capital is limited
- Run quarterly Monte Carlo stress tests
- Set alerts for worst-case scenarios
