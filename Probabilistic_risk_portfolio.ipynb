```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = 10000

clients = np.random.normal(50, 10, n)
avg_fee = np.random.normal(200, 50, n)

revenue = clients * avg_fee

mean_rev = np.mean(revenue)
std_rev = np.std(revenue)
p5_rev = np.percentile(revenue, 5)
p95_rev = np.percentile(revenue, 95)

mean_rev, std_rev, p5_rev, p95_rev
plt.figure()
plt.hist(revenue, bins=50)
plt.axvline(mean_rev)
plt.axvline(p5_rev)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
losses = np.random.normal(100000, 20000, n)

mean_loss = np.mean(losses)
p95_loss = np.percentile(losses, 95)

mean_loss, p95_loss
plt.figure()
plt.hist(losses, bins=50)
plt.axvline(mean_loss)
plt.axvline(p95_loss)
plt.title("Operational Loss Distribution")
plt.xlabel("Loss")
plt.ylabel("Frequency")
plt.show()
strategy_A = np.random.normal(10, 5, n)
strategy_B = np.random.normal(15, 15, n)

mean_A = np.mean(strategy_A)
mean_B = np.mean(strategy_B)

std_A = np.std(strategy_A)
std_B = np.std(strategy_B)

prob_neg_A = np.mean(strategy_A < 0)
prob_neg_B = np.mean(strategy_B < 0)

p5_A = np.percentile(strategy_A, 5)
p5_B = np.percentile(strategy_B, 5)

mean_A, mean_B, prob_neg_A, prob_neg_B, p5_A, p5_B
clients_stress = clients * 0.8
revenue_stress = clients_stress * avg_fee

mean_stress = np.mean(revenue_stress)
p5_stress = np.percentile(revenue_stress, 5)

mean_stress, p5_stress
```
